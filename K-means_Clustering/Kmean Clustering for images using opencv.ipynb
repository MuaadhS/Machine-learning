```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat-animals-tiger-albino-wallpaper-preview.jpg')
```

```python
img.shape
```

```python
plt.figure(figsize = (15,7))
plt.imshow(img)
```

```python
plt.hist(img.flat, bins = 25)
```

```python
#MxNx3 becomes kx3
img2 = img.reshape((-1,3)) # -1 is MxN
img2.shape
```

```python
#opencv requires float
img2 = np.float32(img2)
```

Define criteria, number of clusters and apply k-means<br>
When this criterion is satisfied, the algorithm iteration stops.<br> 
cv.TERM_CRITERIA_EPS  stop the algorithm iteration if specified accuracy, epsilon, is reached.<br>
cv.TERM_CRITERIA_MAX_ITER  stop the algorithm after the specified number of iterations, max_iter.<br>
cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER  stop the iteration when any of the above condition is met.<br>
Max iterations, in this example 10. <br>
Epsilon, required accuracy, in this example 1.0

```python
criteria = (cv2.TERM_CRITERIA_EPS +  cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

#number of clusters
k = 4

attempts = 10
```

```python
ret,label,center=cv2.kmeans(img2, k, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)
```

```python
center = np.uint8(center) 
res = center[label.flatten()]
res2 = res.reshape((img.shape))
```

```python
plt.figure(figsize = (15,7))
plt.imshow(res2)
```

```python
plt.hist(res2.flat, bins = 25)
```
